As a next step, we will apply the apriori algorithm to this data. There are two settings that you probably need to control:

min_support: this is the support threshold that was explained in the algorithm section above. A small difference is that it is expressed as a percentage here rather than a number.
min_confidence: this is merely a filter that filters out rules that do not meet minimum confidence. You can put it to zero if you want to see all the generated rules.

# Using the arl_recommender function, make a service recommendation to a user that uses one service and then what other may be useful

```python
def arl_recommender(rules_df, product_id, rec_count=1):
    sorted_rules = rules_df.sort_values("lift", ascending=False)    # we rank the rules that we have calculated before, according to the lift.
    recommendation_list = []        # Since we can make more than one suggestion, we have created a suggestion list.
    for i, product in enumerate(sorted_rules["antecedents"]):
        for j in list(product):         # We convert the values we have captured into a list.
            if j == product_id:
                recommendation_list.append(list(sorted_rules.iloc[i]["consequents"]))
    return recommendation_list[0:rec_count]
```

# the interpretation  when you do association is that it does not manner how many items of coffe you buy in one transactions it only matter whether it appears or not, if it is bought. The interpretation is that coffee appears in 47% of the transactions and bread in 32% of all transactions. 

# If we want to find the support value of a pair of antecedent and consequent we can add them as follows.

```python
import numpy as np
```

```python
# Add antecedent and consequent pairs to the data set 
# it is important the order here because the order way would give a different result. In this example is first 
# buy Coffe and then the other item. 
basket['Coffee & Bread'] = np.logical_and(basket['Coffee'], basket['Bread'])
basket['Coffee & Tea'] = np.logical_and(basket['Coffee'], basket['Tea'])
basket['Coffee & Cake'] = np.logical_and(basket['Coffee'], basket['Cake'])
```

```python
# Calculate support
support = basket.mean().sort_values(ascending=False)
support.head(20)
```

For example, if we have 10 transactions and 6 of them include coffee and 4 of them include both coffee and cookie, then support of {coffee} is 60% and support of {coffee, cookie} is 40%. 

```python
# Confidence value of if Coffee then Bread 
print(support['Coffee & Bread']/support['Coffee'])

# Confidence value of if Coffee then Cake 
print(support['Coffee & Cake']/support['Coffee'])
```

We can interpret these results as 18% confidence means that 18% of the transactions that include Coffee also include Bread.

It is important to mention that, confidence metric is not symmetric and Confidence(A→B) is different than Confidence(B→A).

```python
# Lift value of if Coffee then Bread 
print(support['Coffee & Bread']/(support['Coffee']*support['Bread']))

# Lift value of if Coffee then Cake 
print(support['Coffee & Cake']/(support['Coffee']*support['Cake']))
```

0.57 for Coffee → Bread
1.10 for Coffee → Cake
As you can see having Coffee in a transaction decreases the chances of having Bread 
while increasing the chances of having a Cake.

```python
# Conviction value of if Coffee then Bread 
print(support['Coffee']*(
        1-support['Bread'])/(support['Coffee'] - support['Coffee & Bread']))

# Conviction value of if Coffee then Cake 
print(support['Coffee']*(
        1-support['Cake'])/(support['Coffee'] - support['Coffee & Cake']))
```

0.82 for Coffee → Bread
1.01 for Coffee → Cake
high is > 1. Bread is not such dependent on Coffe but Cake yes.

# To make use of the apriori module given by mlxtend library, we need to convert the dataset according to it’s liking. apriori module requires a dataframe that has either 0 and 1 or True and False as data. The data we have is all string (name of items), we need to use One Hot Encode the data.Custom One Hot Encoding: 

```python
itemset = set(items)
encoded_vals = []
for index, row in df.iterrows():
    rowset = set(row) 
    labels = {}
    uncommons = list(itemset - rowset)
    commons = list(itemset.intersection(rowset))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
ohe_df = pd.DataFrame(encoded_vals)
```

```python
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
freq_items = apriori(ohe_df2, min_support=0.05, use_colnames=True, verbose=1)
freq_items
```

# Metric can be set to confidence, lift, support, leverage and conviction.

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules.head()
```

```python
rules_2 = association_rules(freq_items, metric="lift", min_threshold=1)
rules_2.head()
```

```python
rules_2.sort_values(by=['lift','confidence'],ascending=False)
```

```python
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(rules['support'], rules['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show()
```

```python
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'], 
fit_fn(rules['lift']))
```
